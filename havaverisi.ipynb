```python
import requests
import pandas as pd

# API URL'si
url = "https://archive-api.open-meteo.com/v1/archive"

# Parametreler (İstanbul koordinatları)
params = {
    "latitude": 41.0082,
    "longitude": 28.9784,
    "start_date": "2024-01-01",
    "end_date": "2024-12-31",
    "hourly": [
        "temperature_2m",
        "precipitation",
        "windspeed_10m",
        "relative_humidity_2m",
        "cloudcover"
    ],
    "timezone": "Europe/Istanbul"
}

# API isteği gönder
response = requests.get(url, params=params)
data = response.json()

# JSON'u pandas DataFrame'e dönüştür
hava_df = pd.DataFrame({
    "Tarih_Saat": data["hourly"]["time"],
    "Sicaklik": data["hourly"]["temperature_2m"],
    "Yagis": data["hourly"]["precipitation"],
    "Ruzgar_Hizi": data["hourly"]["windspeed_10m"],
    "Nem": data["hourly"]["relative_humidity_2m"],
    "Bulutluluk": data["hourly"]["cloudcover"]
})

# CSV'ye kaydet
hava_df.to_csv("istanbul_hava_durumu_saatlik_2024.csv", index=False)

hava_df.head()
```
